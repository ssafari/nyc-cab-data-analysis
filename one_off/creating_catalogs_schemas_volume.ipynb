```python
import urllib.request
import shutil
import os
from urllib.error import URLError, HTTPError
```

## CREATE CATALOG

```python
spark.sql("create catalog if not exists nyctaxi managed location 'abfss://unity-catalog-storage@dbstorage3qjdm3fm2mn5m.dfs.core.windows.net/1065847465169975'")
```

### Create Schemas

```python
spark.sql("create schema if not exists nyctaxi.00_landing")
spark.sql("create schema if not exists nyctaxi.01_bronze")
spark.sql("create schema if not exists nyctaxi.02_silver")
spark.sql("create schema if not exists nyctaxi.03_gold")
```

```python
spark.sql("create volume if not exists nyctaxi.00_landing.data_sources")
```

### Create Volumes

Import data from Yellow Taxi site into Volumes data resources directory

```python
dates_to_process = ['2025-01', '2025-02', '2025-03', '2025-04', '2025-05', '2025-06', '2025-07', '2025-08', '2025-09', '2025-10', '2025-11', '2025-12']

for date in dates_to_process:
    url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_{date}.parquet"
    try:
        response = urllib.request.urlopen(url)
        dir_path = f"/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow/{date}"
        os.makedirs(dir_path, exist_ok=True)
        local_path = f"{dir_path}/yellow_tripdata_{date}.parquet"
        with open(local_path, 'wb') as f:
            shutil.copyfileobj(response, f)
    except HTTPError as e:
        print(f"HTTP Error: {e.code} - {e.reason}")
    except URLError as e:
        print(f"URL Error: {e.reason}")
```

```python
url = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
response = urllib.request.urlopen(url)
dir_path = "/Volumes/nyctaxi/00_landing/data_sources/lookup"
os.makedirs(dir_path, exist_ok=True)
local_path = f"{dir_path}/taxi_zone_lookup.csv"
with open(local_path, 'wb') as f:
    shutil.copyfileobj(response, f)
```
