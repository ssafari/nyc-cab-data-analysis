# Databricks Notebook Yellow Taxi Data Source

```python
from pyspark.sql.functions import current_timestamp
from dateutil.relativedelta import relativedelta
```

```python
# Obtains the year-month for 1 months prior to the current month in yyyy-MM format
one_months_ago = date.today() - relativedelta(months=1)
formatted_date = one_months_ago.strftime("%Y-%m")

# Read all Parquet files from the landing directory into a DataFrame
df = spark.read.format("parquet").load(f"/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow/{formatted_date}")
```

```python
# Add a column to capture when the data was processed
df = df.withColumn("processed_timestamp", current_timestamp())
```

```python
# Write the DataFrame to a Unity Catalog managed Delta table in the bronze schema, overwriting any existing data
df.write.mode("append").saveAsTable("nyctaxi.01_bronze.yellow_trips_raw")
```
